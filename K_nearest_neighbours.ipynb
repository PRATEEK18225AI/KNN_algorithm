```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#defining distance function
def dist(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))

#defining knn algo to return most likely label
def knn(X,Y,x,k=5):
    '''this function take proprety(X),label(Y),query_point(x), nearest neighbour(k, by default 5 as arguments) and it returns a 2d numpy array which distance and label as elements'''
    ll=[]
    for i in range(X.shape[0]):
        d=dist(x,X[i])
        ll.append([d,Y[i]])
    ll=np.array(sorted(ll))
    ll=ll[0:k,:]
    a,b=np.unique(ll[:,1], return_counts=True)
    idx=np.argmax(b)
    return a[idx]
```

### Loading Iris Datasets
https://www.kaggle.com/uciml/iris

```python
df=pd.read_csv('Iris.csv')
df.shape
```

```python
df.head()
```

```python
df.drop(columns=['Id'],inplace=True)
```

```python
#converting into numpy array
X,y=df[df.columns[:-1]].values,df[df.columns[-1]].values
```

### Normalizing The Dataset

```python
mean=X.mean(axis=0)
std=X.std(axis=0)
X=(X-mean)/std
```

### Train Test Split

```python
X_train,X_test,y_train,y_test= train_test_split(X, y, test_size=1/3, random_state=42)
```

```python
print(X_test.shape,y_test.shape)
```

### Testing Samples

```python
predictions=[]
for i in range(50):
    pred=knn(X_train,y_train,X_test[i])
    predictions.append(pred)
    
predictions=np.array(predictions)
acc=(y_test==predictions).sum()/(y_test.shape[0])
print(f'Test Accracy={acc}\n',metrics.classification_report(y_test,predictions))
```

#### Confusion Matrix

```python
def createCFM(y_test, predictions):
    col=np.unique(y_test)
    mat=metrics.confusion_matrix(y_test, predictions)
    data={col[n]:mat.T[n] for n in range(len(col))}
    df=pd.DataFrame(data)
    df.index=[i+'-Real' for i in col]
    return df
```

```python
createCFM(y_test, predictions)
```
